```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd

# Inisialisasi driver 
driver = webdriver.Chrome()

driver.get("https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue")  

# Tunggu beberapa saat agar JavaScript selesai dimuat
driver.implicitly_wait(10)

# Ambil konten HTML setelah halaman dimuat
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

tables = soup.find_all("table")
# Mencari indeks tabel yang mengandung "Tesla Quarterly Revenue"
table_index = None
for index, table in enumerate(tables):
    if "Tesla Quarterly Revenue" in str(table):
        table_index = index
        break  # Menghentikan loop setelah menemukan tabel yang sesuai

# Memeriksa apakah table_index terdefinisi
if table_index is not None:
    Tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])

    for row in tables[table_index].tbody.find_all("tr"):
        col = row.find_all("td")
        if col != []:
            Date = col[0].text
            Revenue = col[1].text.replace("$", "").replace(",", "")
            # Membuat DataFrame sementara untuk setiap baris dan menggabungkannya
            temp_df = pd.DataFrame({"Date": [Date], "Revenue": [Revenue]})
            Tesla_revenue = pd.concat([Tesla_revenue, temp_df], ignore_index=True)

    # Menampilkan hasil akhir
    print(Tesla_revenue)
else:
    print("Tabel 'Tesla Quarterly Revenue' tidak ditemukan.")
```

```python
Tesla_revenue = Tesla_revenue[Tesla_revenue['Revenue'] != ""]
```

## Point 4 : Use Webscraping to Extract GME Revenue Data

```python
# inisialisasi driver 
driver = webdriver.Chrome()
driver.get("https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue")  # Replace with actual URL

# mennuggu javascript loading data
driver.implicitly_wait(10)

# mengekstak halaman data dan parse dengan BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

# menemukan tabel 
tables = soup.find_all("table")
table_index = None
for index, table in enumerate(tables):
    if "GameStop Quarterly Revenue" in str(table):
        table_index = index
        break

# inisialisasi dataframe
gme_revenue = pd.DataFrame(columns=["Date", "Revenue"])

if table_index is not None:
    for row in tables[table_index].tbody.find_all("tr"):
        col = row.find_all("td")
        if col != []:
            Date = col[0].text
            Revenue = col[1].text.replace("$", "").replace(",", "")
            gme_revenue = pd.concat([gme_revenue, pd.DataFrame({"Date": [Date], "Revenue": [Revenue]})], ignore_index=True)

# menampilkan 5 baris terkahir
gme_revenue.tail()
```

## Point 5: Plot Tesla Stock Graph

```python
import matplotlib.pyplot as plt

# Plot Tesla stock data
def make_graph(df, title="Stock Data"):
    df['Close'].plot(figsize=(10, 6), title=title)
    plt.xlabel('Date')
    plt.ylabel('Close Price ($)')
    plt.grid(True)
    plt.show()

# Assuming Tesla stock data (Tesla_revenue) is loaded as a DataFrame
make_graph(gme_data, title="Tesla Stock Data")
```
